```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/workspaces/DS522---Assignments/sales_data_sample.csv', encoding='latin1')
df
```

```python
# Checking for missing values
print(df.isnull().sum())

# Checking data types and basic information
print(df.info())
```

```python
print(df.describe())
```

```python
# Drop all rows with missing values
df_cleaned = df.dropna()

# Display the cleaned DataFrame
df_cleaned
```

```python
# Drop columns with missing values
df_cleaned = df.dropna(axis=1)
df_cleaned
```

```python
# Make a copy of the df
df1 = df.copy()

# Imputing missing values of the copy of df:
# Filling with mean
df1['PRICEEACH'] = df1['PRICEEACH'].fillna(df1['PRICEEACH'].mean())

# Checking for missing values
print(df1.isnull().sum())
print(df1.info())
```

```python
# Filling with the column's mode value
mode_val = df1['DEALSIZE'].mode()
print(mode_val)
df1['DEALSIZE'] = df1['DEALSIZE'].fillna(mode_val)
```

```python
# Fill missing data with the previous value
df1['POSTALCODE'].ffill(inplace=True)

# Checking for missing values
print(df1.isnull().sum())
```

```python
# Checks rows for duplicates in the entire dataframe
# Returns a boolean Series where True indicates a duplicate row
duplicates = df1.duplicated()

# To check the total number of duplicate rows
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
# To display the duplicate rows themselves
duplicate_rows = df1[df1.duplicated()]
print(duplicate_rows)
```

```python
# Drop duplicate rows (keeping the first occurrence)
df_cleaned = df1.drop_duplicates()

# Checks rows for duplicates in the entire dataframe
duplicates = df_cleaned.duplicated()

# To check the total number of duplicate rows
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
# Drop duplicates based on specific columns
df_cleaned = df1.drop_duplicates(subset=['ORDERNUMBER', 'PRODUCTCODE'])

# Checks rows for duplicates in the entire dataframe
duplicates = df_cleaned.duplicated()

# To check the total number of duplicate rows
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
# Compute the first quartile of the SALES column
Q1 = df1['SALES'].quantile(0.25)

# Compute the third quartile of the SALES column
Q3 = df1['SALES'].quantile(0.75)

# Compute the interquartile range (IQR)
IQR = Q3 - Q1

# Remove outliers using IQR rule
df_cleaned = df1[~((df1['SALES'] < (Q1 - 1.5 * IQR)) | (df1['SALES'] > (Q3 + 1.5 * IQR)))]
df_cleaned
```

```python
# Rename the specified column names
df_cleaned.rename(columns={
    'ORDERNUMBER': 'Order_Number',
    'QUANTITYORDERED': 'Quantity_Ordered',
    'PRICEEACH': 'Price_Each',
    'ORDERLINENUMBER': 'Order_Line_Number',
    'SALES': 'Sales',
    'ORDERDATE': 'Order_Date',
    'QTR_ID': 'Quarter',
    'MONTH_ID': 'Month',
    'YEAR_ID': 'Year',
    'PRODUCTLINE': 'Product_Line',
    'PRODUCTCODE': 'Product_Code',
    'CUSTOMERNAME': 'Customer_Name',
    'PHONE': 'Phone',
    'ADDRESSLINE1': 'Address',
    'CITY': 'City',
    'STATE': 'State',
    'POSTALCODE': 'Postal_Code',
    'COUNTRY': 'Country',
    'CONTACTLASTNAME': 'Contact_Last_Name',
    'CONTACTFIRSTNAME': 'Contact_First_Name',
    'DEALSIZE': 'Deal_Size'
}, inplace=True)

df_cleaned.head()
```
